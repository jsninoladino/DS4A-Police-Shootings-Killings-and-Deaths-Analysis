```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
from bs4 import BeautifulSoup
import urllib
```

```python
# Open a file and pass the file handle (here file handle is f) to BeautifulSoup
# html file was saved before hand
file_name = 'C:/Users/Johnny/Documents/Classes/DS4A/Project/HTML/saved_page2000.html'
with open(file_name) as f:  #Windows users may need to add the option encoding='utf8'
    officers_2020 = BeautifulSoup(f, 'html.parser')
```

```python
# finds all officer details from webpage and adds it to a list
case_n = []
for i in officers_2020.find_all('div',{'class':'officer-short-details'}):
    case_n.append(i.text)
```

```python
case_n
```

```python
# make lists to separate the gathered information in case_n
police_name = []
agency = []
eow = []
cause = []


for case in case_n:
    police_name.append(case.split('\n')[1])
    agency.append(case.split('\n')[2])
    eow.append(case.split('\n')[3])
    cause.append(case.split('\n')[4].split(': ')[1])
```

```python
# make lists to separate the gathered information in case_n
dept = []
state = []
for i in agency:
    dept.append(i.split(', ')[0])
    state.append(i.split(', ')[1])
```

```python
# make lists to separate the gathered information in case_n
month_day = []
year = []
day_of_week = []

for i in eow:
    month_day.append(i.split(', ')[1])
    year.append(i.split(', ')[2])
    day_of_week.append(i.split(', ')[0].split(': ')[1])
```

```python
#convert to a dataframe

police_omdb2020 = pd.DataFrame({'police_name':police_name,
                                'dept':dept,
                                'state':state,
                                'cause':cause,
                                'month_day':month_day,
                                'year':year,
                                'day_of_week':day_of_week})
```

```python
# repeat steps from above but with the 20 html files
officers_years = []
for i in range(0,len(years_int)):
    # Open a file and pass the file handle (here file handle is f) to BeautifulSoup
    file_name = 'saved_page'+str(years_int[i])+'.html'
    with open(file_name) as f:  #Windows users may need to add the option encoding='utf8'
        officers_years.append(BeautifulSoup(f, 'html.parser'))
```

```python
police_omdbdfs = []

for officer_year in officers_years:
    
    case_n = []
    for i in officer_year.find_all('div',{'class':'officer-short-details'}):
        case_n.append(i.text)

    police_name = []
    agency = []
    eow = []
    cause = []


    for case in case_n:
        police_name.append(case.split('\n')[1])
        agency.append(case.split('\n')[2])
        eow.append(case.split('\n')[3])
        cause.append(case.split('\n')[4].split(': ')[1])

    dept = []
    state = []
    for i in agency:
        dept.append(i.split(', ')[0])
        state.append(i.split(', ')[1])

    month_day = []
    year = []
    day_of_week = []

    for i in eow:
        month_day.append(i.split(', ')[1])
        year.append(i.split(', ')[2])
        day_of_week.append(i.split(', ')[0].split(': ')[1])

    police_omdbdfs.append(pd.DataFrame({'police_name':police_name,
                                        'dept':dept,
                                        'state':state,
                                        'cause':cause,
                                        'month_day':month_day,
                                        'year':year,
                                        'day_of_week':day_of_week}))
```

```python
police_omdb = pd.concat(police_omdbdfs).reset_index(drop=True)
```
